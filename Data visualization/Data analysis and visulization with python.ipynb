```python
# using functions as converters.

def convert_people_cell(cell):
    if cell == 'n.a.':
        return 'Sam Walton'
    return cell

def convert_eps_cell(cell):
    if cell == "not available":
        return None
    return cell

df = pd.read_excel("sample_data.xlsx","Sheet1",header=1,converters={
    'people' : convert_people_cell,
    'eps' : convert_eps_cell
})

df
```

## Handling missing values

```python
df = pd.read_csv("missing_value_data.csv")
df.head()
```

```python
# type of date
type(df.Day[0])
```

```python
# to get date in timestamp of python.
df = pd.read_csv("missing_value_data.csv",parse_dates=["Day"])
df.head()
```

```python
type(df.Day[0])
```

```python
df.set_index("Day",inplace=True)
```

```python
df.head()
```

```python
# our df has many missing values
# lets replace missing values to 0(Zero).
new_df = df.fillna(0)
new_df.head()
```

```python
# if we want to replace data by column specifically then..
new_df = df.fillna({
    'temperature' : 0,
    'windspeed' : 0,
    'event' : 'no event'
})
new_df
```

```python
# ffill means forward fill, means it will fill missing values by value exactly above it.
# ffill and bfill (backward fill) works rowwise or vertically by default.
new_df = df.fillna(method = 'ffill')
# now it will fill missing value by it's above value
new_df
```

```python
new_df = df.fillna(method = 'bfill')
# now it will fill missing value by it's below value
new_df
```

```python
# if we want to use ffill & bfill horizontally then..
new_df = df.fillna(method = 'ffill',axis="columns")
# now it will fill missing value by it's left side value
# if it get left side index column, then that value will NaN, because we index values are not accessible.
new_df
```

```python
new_df = df.fillna(method = 'bfill',axis="columns")
# now it will fill missing value by it's right side value
new_df
```

```python
# if we want to fill missing values only specific times continuously, then provide limit
new_df = df.fillna(method = 'bfill',limit=1)
# it will missing value only one time everywhere, after then continuously if missing value found 
# then it will put NaN
new_df
```

```python
new_df = df.fillna(method = 'bfill',limit=2)
# it will missing value maximum two time everywhere, after then continuously if missing value found 
# then it will put NaN
new_df
```

```python
# now look at our original data
df.head()
```

```python
# in this data, second row's temperature is not available.
# we can assume that that value should be in between on it's above and below value.
# so here it should be (32+28)/2 = 30
# this is called linear interpolation
```

```python
# lets use linear interpolation
new_df = df.interpolate()
new_df
```

```python
# but still here there is a problem.
# second row's temperature was missing.
# that value should near to it's below value.
# because look at those row's Day.
# 5th month is more near to 4th month then 1st month.
# so that temperature should near to 28.
# so now we have to interpolate data by time.
```

```python
# interpolate data by time method.
new_df = df.interpolate(method = "time")
new_df
```

```python
# to drop every rows with missing values
new_df = df.dropna()
new_df
```

```python
# here, how = all means which row has every cell with missing value, drop that row only
new_df = df.dropna(how="all")
new_df
```

```python
# thresh means threshold
# this statement will keep all the rows which have atleast 1 non NaN value.
new_df = df.dropna(thresh=1)
new_df
```

```python
# this statement will keep all the rows which have atleast 2 non NaN value.
new_df = df.dropna(thresh=2)
new_df
```

```python
# this statement will keep all the rows which have atleast 3 non NaN value.
new_df = df.dropna(thresh=3)
new_df
```

```python
import pickle

# lets do the above program again in functional manner
def state_list(): # for return fiddy states like above
    url = "https://simple.wikipedia.org/wiki/List_of_U.S._states" # creating url
    html = request.urlopen(url, context=ssl._create_unverified_context()).read()
    fiddy_states = pd.read_html(html) # returns list of dataframes
    return fidd_abbv.Abbreviation

def grab_initial_state_data(): # for return the states initial data like above
    states = state_list()
    main_df = pd.DataFrame() # making the df for use below
    for abbv in states:
        df = quandl.get(str("FMAC/HPI_"+str(abbv)),authtoken = "FvtsM3v25sS2N3wuyDh_")
        if main_df.empty:
            main_df = df
        else:
            main_df = main_df.append(df)
    print(main_df)
    # writing everything to pickle bytes buffer file
    pickle_out = open("fiddy_states.pickle","wb") # creating pickle file & open it in write bytes mode.
    pickle.dump(main_df,pickle_out) # writing to file
    pickle_out.close() # closing file
```

```python
# writing the above functional manner program again
# lets use percentage-change in here

import pickle

# lets do the above program again in functional manner
def state_list(): # for return fiddy states like above
    url = "https://simple.wikipedia.org/wiki/List_of_U.S._states" # creating url
    html = request.urlopen(url, context=ssl._create_unverified_context()).read()
    fiddy_states = pd.read_html(html) # returns list of dataframes
    return fidd_abbv.Abbreviation

def grab_initial_state_data(): # for return the states initial data like above
    states = state_list()
    main_df = pd.DataFrame() # making the df for use below
    for abbv in states:
        df = quandl.get(str("FMAC/HPI_"+str(abbv)),authtoken = "FvtsM3v25sS2N3wuyDh_")
        df = df.pct_change() # add this for find parcentage change
        if main_df.empty:
            main_df = df
        else:
            main_df = main_df.append(df)
    print(main_df)
    # writing everything to pickle bytes buffer file
    pickle_out = open("fiddy_states.pickle","wb") # creating pickle file & open it in write bytes mode.
    pickle.dump(main_df,pickle_out) # writing to file
    pickle_out.close() # closing file
```
