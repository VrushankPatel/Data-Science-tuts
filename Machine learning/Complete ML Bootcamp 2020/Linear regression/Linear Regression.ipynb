```python
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("cost_revenue_clean.csv")
```

```python
data.describe()
```

```python
5.034000e+03
```

```python
x = DataFrame(data, columns=['production_budget_usd'])
y = DataFrame(data, columns=['worldwide_gross_usd'])
```

```python
x.describe()
```

```python
y.describe()
```

```python
x.head
```

```python
y.head
```

```python
plt.figure(figsize=(10,6)) # size of chart
plt.scatter(x,y)  # scatter plot
plt.title('Film costs vs global revenue') # title of chart
plt.xlabel('Production budget $')  # x axis label
plt.ylabel('worldwide gross $')   # y axis label
plt.show()   # request to show the plot
```

### in above plot, there is some space left at the starting of axises and points
### to overcome this, we should apply limits to the points of axis
#### here, the y axis (vertical) points limit seems to be between 0 and 3 bilion.
#### where, the x axis (horizontal) points limit seems to be between 0 and 450 million.


```python
# same plot as above, but with applying the limits
plt.figure(figsize=(10,6)) # size of chart
plt.scatter(x, y, alpha=0.3)  # scatter plot with dots transperency of 0.3, by default, it was 1
plt.title('Film costs vs global revenue') # title of chart
plt.xlabel('Production budget $')  # x axis label
plt.ylabel('worldwide gross $')   # y axis label
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.show()
```

```python
regression = LinearRegression()
regression.fit(x, y)
```

let's get slope coffecient of linear regression model we just created

```python
regression.coef_      # theta_1
```

in above, as per regression coeficcient value of 3.1111..
it means that for each dollar, 3.11150918 dollar revenue is generated.

let's get intersept coffecient of linear regression model we just created

```python
regression.intercept_
```

```python
# same plot as above, but with regression line
plt.figure(figsize=(10,6)) # size of chart
plt.plot(x, regression.predict(x), color='red', linewidth=4) # line of points of prediction for x column

plt.scatter(x, y, alpha=0.3)  # scatter plot with dots transperency of 0.3, by default, it was 1
plt.title('Film costs vs global revenue') # title of chart
plt.xlabel('Production budget $')  # x axis label
plt.ylabel('worldwide gross $')   # y axis label
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.show()
```

let's check our regression model's efficiency by score

```python
regression.score(x, y)
```

which means, our algorithm is 54.96 percent accurate.
